```python
#Tasca M3 T02   Enric Sena Alvarez
#Exericici 1
import pandas as pd
import numpy as np

data=pd.read_csv("DelayedFlights.csv")
```

```python
data.shape #Veig les dimensions del dataset original
data=data.sample(n=200000) #redueixo la dimensi√≥ escollint aleatoriament 200.000 files. (sense repeticions)
```

```python
print(data.columns)
data[:20]

#Escullo les columnes rellevants.
dades=data[["UniqueCarrier","DepTime","ArrTime","AirTime","Origin","Dest","Distance","CarrierDelay","FlightNum"]]
```

```python
dades[:10] #comprovo que s'hagi fet el subset correctament.
```

```python
#Resum estad√≠stic de les columnes rellevants:
dades.describe()

#nombre de na's per columnes
dades.isnull().sum(axis = 0)
dades=pd.DataFrame(dades)
dades['AirTimeH']=dades['AirTime']/60 #passo de minuts a hores
dades['DistKM']=dades['Distance']*1.609344 #Passo de milles a quil√≤metres

#Calculo la velocitat mitjana per vol:
dades['Vel']=dades['DistKM']/dades['AirTimeH']
dades[:30]
```

```python
#faig una taula amb les aerolinies amb m√©s retards:
retards=dades.sort_values(by=["CarrierDelay"],ascending=False)
retards[:20]
retards.CarrierDelay.quantile(0.9999) #obtinc el quantil 99.99% i aix√≠ puc seleccionar les aerolinies amb un retr√†s major a aquest.

aero_retards = retards[retards['CarrierDelay']>879]
aero_retards

aero=aero_retards.groupby("UniqueCarrier")
aero.first()

#Les aerolinies amb m√©s retard s√≥in American Airlines, Copa airlines i North West airlines.
```

```python
# Vols m√©s llargs.
long=dades.sort_values(by=["Distance"],ascending=False)
aero_long=long[long['Distance']>long.Distance.quantile(0.999)]
aero_long
aero_long_ord=aero_long.groupby("FlightNum")
aero_long_ord.first().sort_values(by=["Distance"])

#Els vols m√©s llargs s√≥n: NY-Honolulu, Atlanta-Honolulu, Honolulu-Chicago
```

```python
#Vols m√©s endarrerits:
retards[:20]
dest_retard= retards[retards['CarrierDelay']>retards.CarrierDelay.quantile(0.9999)]
dest_retard[:20]
dest_retards=dest_retard.groupby("FlightNum")
dest_retards.first().sort_values(by=["CarrierDelay"],ascending=False)

#Els vols amb m√©s retard s√≥n els que tenen com a numero de vol: 776,1515,1192,522,1416,336,1432,2012,6,882,377.
```

```python
#rutes m√©s llargues amb m√©s retard

dest_retards.first().sort_values(["Distance","CarrierDelay"],ascending=False)
#Els vols amb m√©s retard i amb rutes m√©s llargues s√≥n: 6,1192,336,522,1416,2012....
```
