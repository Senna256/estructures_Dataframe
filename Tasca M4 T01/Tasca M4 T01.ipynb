```python
#Tasca 4 Enric Sena Alvarez

#Exercici 1
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#lloguer=pd.read_csv("preu_lloguer.csv")
lloguer = pd.read_csv('preu_lloguer.csv',na_values = '--')
lloguer.head()

grafic=sns.displot(data=lloguer, x="Nom_Districte")
grafic.set_xticklabels(rotation = 90)
```

```python
lloguer.dtypes
preus = lloguer[lloguer.Preu.notnull()] #en principi aix√≤ hauria d'eliminar els zeros i per tant fer l'histograma sense NA's
```

```python
grafic_preu=sns.distplot( a=preus["Preu"], hist=True, kde=True, rug=False )
```

```python
grafic2=sns.catplot(x="Nom_Districte", y="Preu", data=lloguer)
grafic2.set_xticklabels(rotation = 90)
```

```python
grafic3=sns.relplot(x="Trimestre", y="Preu", data=lloguer)
grafic3
```

```python
grafic4=sns.relplot(x="Trimestre", y="Preu", hue="Nom_Districte", data=lloguer)
```
